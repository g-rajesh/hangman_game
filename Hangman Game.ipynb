```python
import random
import re
from IPython.display import clear_output
global playing
```

```python
def hang(hangman):
        return hangman.pop(0)
    
```

```python
def guess():
    words = ['rainbow','lion','mango','tajmahal','school','fridge','maths','samantha','hardwork','undertaker']
    dic_words = {'rainbow':'color','lion':'king','mango':'fruit','tajmahal':'love','school':'memories',
                 'fridge':'iceland','maths':'subject','samantha':'theri','hardwork':'success','undertaker':'phenom'}

    word = random.choice(words).upper()
    dic_word = dic_words[word.lower()]
    return list(word),dic_word.upper(),word
```

```python
def name():
    while True:
        try:
            player_name = input('PLEASE INPUT YOUR NAME : ').upper()
            player_name = ''.join(re.findall(r'[^.!?,0-9 ]',player_name))
        except:
            print('Only letters are allowed')
            continue
        else:
            break
    return player_name
```

```python
def play():
    while True:
        try:
            play_or_not = input('Do you want to play the game(y/n) : ')
        except:
            print('Please enter either yes or no or (y/n)')
            continue
        else:
            if play_or_not[0].lower() == 'y':
                return True
            elif play_or_not[0].lower() == 'n':
                return False
            else:
                print('Please enter either yes or no or (y/n)')
                continue
            
        
```

```python
def ready_or_not():
    while True:
        try:
            ready = input('Ready to play? (y/n) : ')
        except:
            print('Please input yes or no or (y/n)')
        else:
            if ready[0].lower() == 'y':
                return True
            elif ready[0].lower() == 'n':
                return False
            else:
                print('Please input yes or no or (y/n)')
                continue
        
```

```python
def play_again():
    while True:
        try:
            ready = input('Play again? (y/n) : ')
        except:
            print('Please input yes or no or (y/n)')
        else:
            if ready[0].lower() == 'y':
                return True
            elif ready[0].lower() == 'n':
                return False
            else:
                print('Please input yes or no or (y/n)')
                continue
    
```

```python
global playing

hanged_man = ''

print('Welcome to Hangman Game!!!')
print('\n'*2)

player_name = name()
playing = play()

while playing:
    hangman = [
    '\n    ________\n   |       |\n   |        \n   |\n   |\n   |\n   |',
    '\n    ________\n   |       |\n   |       0\n   |\n   |\n   |\n   |',
    '\n    ________\n   |       |\n   |       0\n   |       | \n   |       \n   |     \n   |',
    '\n    ________\n   |       |\n   |       0\n   |       | \n   |       |\n   |     \n   |',
    '\n    ________\n   |       |\n   |       0\n   |     / | \\\n   |       |\n   |  \n   |',
    '\n    ________\n   |       |\n   |       0\n   |     / | \\\n   |       |\n   |     /   \\\n   |'
    ]
    print('\n')
    print('-------------------------------------------------------------------------------------------')
    print('Instructions:')
    print('1. This is a guessing game. Guess a letter or a word')
    print('2. If you guess all the letters of the word correctly, then you can save the hangman')
    print('3. If you guess wrong letter or word for the sixth time, then the man will be hanged')
    print('BE COOL AND PLAY THE GAME')
    print('-------------------------------------------------------------------------------------------')
    print('\n')
    
    ready = ready_or_not()
    
    guessing_word,hint,check_word = guess()
    guessed = ['_']*len(guessing_word)
    
    while ready:
        
        your_guess = input(f'Guess the letter of the word or word itself.[Hint:{hint.upper()}]').upper()
        count=0
        
        if your_guess == check_word:
            print(f'Congratulations {player_name}! You have saved the man!!!')
            ready = False
            break
            
        elif your_guess in guessing_word:
            indices = list(filter(lambda x:guessing_word[x] == your_guess,range(len(guessing_word))))
            
            for index in indices:
                guessed.pop(index)
                guessed.insert(index,your_guess)
                
            if ''.join(guessed) == check_word:
                print(f'Congratulations {player_name}! You have saved the man!!!')
                ready = False
                break
        else:
            hanged_man = hang(hangman)
            
            if hanged_man == '\n    ________\n   |       |\n   |       0\n   |     / | \\\n   |       |\n   |     /   \\\n   |':
                clear_output()
                print(hanged_man)
                print(f'Sorry {player_name}! The man has been hanged in the hanger!!!')
                ready = False
                break
                
            else:
                count = 1
                
        
        clear_output()
        print('Now, your word '+' '.join(guessed))
        print('\n')
        if count == 1:
            print('Wrong Entry...')
        print('hanging man.....',hanged_man)
        
                

    if  not play_again():
        playing = False
        
    
    
    
print('\n\nFinished playing....')
```
